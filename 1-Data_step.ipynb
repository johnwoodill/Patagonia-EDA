```python
import pandas as pd
import numpy as np
```

```python
dat = pd.read_csv('/data2/GFW_point/2016-01-01/messages-2016-01-01-000000000000.csv')
gfw_dat = pd.read_csv('~/Data/GFW_public/fishing_vessels/fishing_vessels.csv')
```

```python
def data_step(data): 
    # Patagonia shelf coordinates
    #lower left lat: -58
    #lower left lon: -77
    #upper right lat: -23
    #upper right lon: -22
    
    lon1 = -77
    lon2 = -22
    lat1 = -58
    lat2 = -23
    retdat = data[(data['lon'] >= lon1) & (data['lon'] <= lon2) & (data['lat'] >= lat1) & (data['lat'] <= lat2)]
    
    # Separate Year, month, day, hour, minute, second
    retdat['year'] = pd.DatetimeIndex(retdat['timestamp']).year 
    retdat['year'] = retdat.timestamp.dt.to_period('M')
    retdat['month'] = pd.DatetimeIndex(retdat['timestamp']).month
    retdat['day'] = pd.DatetimeIndex(retdat['timestamp']).day
    retdat['hour'] = pd.DatetimeIndex(retdat['timestamp']).hour
    retdat['minute'] = pd.DatetimeIndex(retdat['timestamp']).minute
    retdat['second'] = pd.DatetimeIndex(retdat['timestamp']).second
    
    # Merge GFW ID data
    retdat = pd.merge(retdat, gfw_dat, how='left', on='mmsi')
    
    return retdat
```

```python
ndat = data_step(data=dat)
```
